## Ödev 1: Not Ortalaması Hesaplama (Kolay)
**Zorluk:** ⭐⭐☆☆☆

Kullanıcıdan 5 adet not alın ve bunları bir listede saklayın. Bu notların:
- Ortalamasını
- En yüksek notu
- En düşük notu
- Kaç tane geçer not olduğunu (50 ve üzeri)

hesaplayın ve ekrana yazdırın.

**Örnek Çıktı:**
```
1. notu girin: 85
2. notu girin: 92
3. notu girin: 78
4. notu girin: 95
5. notu girin: 88

Notlar: [85, 92, 78, 95, 88]
Ortalama: 87.6
En yüksek not: 95
En düşük not: 78
Geçer not sayısı: 5
```

**İpuçları:**
- Boş liste oluşturun
- append() ile notları ekleyin
- sum(), max(), min() fonksiyonlarını kullanın
- List comprehension ile geçer notları filtreleyin

````python
# Başlangıç kodu - tamamlayın
print("=== NOT ORTALAMASI HESAPLAMA ===")

# Boş liste oluştur
notlar = [85, 92, 78, 95, 88]

# 5 not al ve listeye ekle
# Kodunuzu buraya yazın...

# Ortalama hesapla
# Kodunuzu buraya yazın...
ortalama = sum(notlar) / len(notlar)
print(f'notlar ortalamasi = {ortalama}')
# En yüksek ve en düşük not
# Kodunuzu buraya yazın...
en_kucuk = min(notlar)
en_buyuk = max(notlar)
print(f'en buyuk not = {en_buyuk}, en kucuk not = {en_kucuk}')

# Geçer not sayısı (50 ve üzeri)
gecer_notlar = []
for puan in notlar:
  if puan >= 50:
    gecer_notlar.append(puan)
print(f'gecer notlar = {gecer_notlar}')
# Kodunuzu buraya yazın...
````

**İpuçları:**
- List comprehension kullanın

## Ödev 4: Liste Manipülasyonu (Orta)
**Zorluk:** ⭐⭐⭐☆☆

- İlk 3 sayıyı silin

**Örnek Çıktı:**
```
1. sayıyı girin: 10

3. sayıyı girin: 30
4. sayıyı girin: 40

İlk 3 sayı silindi: [40, 50, 60]

Sona 3 sayı eklendi: [40, 50, 60, 100, 200, 300]
```

## Ödev 5: Matris İşlemleri (Orta)
**Zorluk:** ⭐⭐⭐☆☆

3x3'lük bir matris oluşturun ve:
- Matrisi ekrana yazdırın
- Ana köşegen elemanlarının toplamını bulun
- Yan köşegen elemanlarının toplamını bulun
- En büyük elemanı bulun

**Örnek Çıktı:**
```
Matris:
[1, 2, 3]
[4, 5, 6]
[7, 8, 9]

Ana köşegen: [1, 5, 9] = 15
Yan köşegen: [3, 5, 7] = 15
En büyük eleman: 9
```

**Gelişmiş versiyon:**
```
Matris:
1  2  3
4  5  6
7  8  9

Ana köşegen toplamı: 15
Yan köşegen toplamı: 15
En büyük eleman: 9 (konum: [2][2])
```

**İpuçları:**
- İç içe liste oluşturun
- İndeks ile köşegen elemanlarına erişin
- max() fonksiyonu ile en büyük elemanı bulun

````python
# Başlangıç kodu - tamamlayın
print("=== MATRİS İŞLEMLERİ ===")

# 3x3 matris oluştur
matris = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Matrisi yazdır
# Kodunuzu buraya yazın...

# Ana köşegen elemanları (sol üstten sağ alta)
# Kodunuzu buraya yazın...
toplam = 0
for i in range(3):
  toplam += matris[i][i]
print(f'ana kosegenler toplami = {toplam}')
# Yan köşegen elemanları (sağ üstten sol alta)
# Kodunuzu buraya yazın...
toplam = 0
for j in range(3):
    for i in range(2, -1, -1): # 2, 1, 0
        toplam += matris[i][j]
print(f'ters kosegenler toplami = {toplam}')

# En büyük elemanı bul
# Kodunuzu buraya yazın...
buyuk_sayi = max(matris[0])
for i in range(1, 3):
    if buyuk_sayi < max(matris[i]):
        buyuk_sayi = max(matris[i])
print(f'en buyuk sayi = {buyuk_sayi}')
````

## Ödev 10: Liste Analizi (Orta)
**Zorluk:** ⭐⭐⭐☆☆

Kullanıcıdan 20 adet sayı alın. Bu sayıların:
- Kaç tanesi asal sayı
- Kaç tanesi mükemmel sayı
- Kaç tanesi palindrom sayı
- Kaç tanesi Armstrong sayısı

olduğunu bulun.

**Örnek Çıktı:**
```
Sayıları girin:
1: 7
2: 28
3: 121
4: 153
5: 13
...

Sayılar: [7, 28, 121, 153, 13, ...]

Analiz sonuçları:
Asal sayılar: 5 adet [7, 13, 17, 19, 23]
Mükemmel sayılar: 1 adet [28]
Palindrom sayılar: 3 adet [121, 131, 141]
Armstrong sayıları: 2 adet [153, 370]
```

**İpuçları:**
- Asal sayı kontrolü için bölen sayısı
- Mükemmel sayı: bölenlerinin toplamı kendisine eşit
- Palindrom: tersi kendisine eşit
- Armstrong: basamaklarının küplerinin toplamı kendisine eşit

````python
# Başlangıç kodu - tamamlayın
print("=== LİSTE ANALİZİ ===")

# Yardımcı fonksiyonlar tanımla
def asal_mi(sayi):
    for i in range(2, sayi):
        if sayi % i == 0:
            return False
    return True

def mukemmel_mi(sayi):
    bolenler = []
    for i in range(1, sayi):
        if sayi % i == 0:
            bolenler.append(i)
    return sum(bolenler) == sayi

def palindrom_mi(sayi):
    return str(sayi)[::-1] == str(sayi)

def armstrong_mi(sayi):
    toplam = 0
    for basamak in str(sayi):
        toplam += int(basamak) ** 3
    return toplam == sayi


# Boş liste oluştur
sayilar = [153, 28, 121, 17, 496, 89, 371, 131, 44, 7, 1331, 23, 6, 407, 191, 13, 8128, 77, 370, 181]

asallar = []
armstrong = []
mukemmel = []
palindrome = []
for sayi in sayilar:
    if asal_mi(sayi):
        asallar.append(sayi)
    if armstrong_mi(sayi):
        armstrong.append(sayi)
    if mukemmel_mi(sayi):
        mukemmel.append(sayi)
    if palindrom_mi(sayi):
        palindrome.append(sayi)
print(asallar)
print(mukemmel)
print(palindrome)
print(armstrong)
# 20 sayı al ve listeye ekle
# Kodunuzu buraya yazın...

# Analiz yap
# Kodunuzu buraya yazın...
````
